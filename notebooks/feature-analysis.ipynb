```python
import numpy as np
import pandas as pd
from collections import defaultdict as ddict
```

```python
case_type = ('cases', 'infections')
region = 'Florida'
region_short = 'fl'
f_ground_truth = f'../data/usa/data_{case_type[0]}_fl.csv'
```

```python
df_region = pd.read_csv(f_ground_truth, index_col='region').transpose()
df_region.index.set_names(['date'], inplace=True)
df_region.index = pd.to_datetime(df_region.index)
df_region = df_region.diff(axis=0)
print('Days = {}, Regions = {}'.format(*df_region.shape))
display(df_region.tail())
```

```python
def load_features(path):
    df = pd.read_csv(path, index_col='region')
    df = df.drop(columns=['type']).transpose()
    df.index.set_names(['date'], inplace=True)
    df.index = pd.to_datetime(df.index)
    return df

dfs = [
    ('wcli state', 'Florida', load_features('../data/usa/symptom-survey/fb-survey_smoothed_wcli-state.csv')),
    ('wcli county', None, load_features('../data/usa/symptom-survey/fb-survey_smoothed_wcli-county.csv')),
    ('cmnty cli state', 'Florida', load_features('../data/usa/symptom-survey/fb-survey_smoothed_hh_cmnty_cli-state.csv')),
    ('cmnty cli county', None, load_features('../data/usa/symptom-survey/fb-survey_smoothed_hh_cmnty_cli-county.csv')),
    ('doctor state', 'Florida', load_features('../data/usa/symptom-survey/doctor-visits_smoothed_adj_cli-state.csv')),
    ('doctor county', None, load_features('../data/usa/symptom-survey/doctor-visits_smoothed_adj_cli-county.csv')),
    ('test ratio', None, load_features('../data/usa/testing/ratio_features_county.csv')),
    ('test total', None, load_features('../data/usa/testing/total_features_county.csv')),
]
display(dfs[-2][2].tail())
```

```python
lagged = {}
lags = [0, 1, 3, 5, 7, 10, 14, 21, 28]
for lag in lags:
    corrs = ddict(list)
    for name, state, df in dfs:
        #df = df.diff(axis=0).fillna(0)
        for county in df_region.columns:
            query = county if state is None else state
            if query not in df.columns:
                corr = np.nan
            else:
                merged = pd.merge_asof(df_region[county], df[query].shift(lag), left_index=True, right_index=True)
                corr = merged.corr().iloc[0, 1]
            corrs[county].append(corr)
    corrs = [(k,) + tuple(v) for k, v in corrs.items()]
    df = pd.DataFrame(corrs, columns=['region'] + [n for n, _, _ in dfs])
    df.set_index('region', inplace=True)
    lagged[lag] = df.mean()
pd.DataFrame(lagged)
```
